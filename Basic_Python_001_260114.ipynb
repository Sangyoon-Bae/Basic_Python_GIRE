### (1) 가장 기본적인 formatting (% 사용)
- d : integer
- f : float
- s : string

```python
print('temperature is %d' % a)
```

```python
b = 3
print('I ate %d apples today.' % b)
```

```python
d = 'dog'
print('I love %s' % d)
```

```python
c = '3'
print('I ate %s apples today.' % c)
```

```python
print('error rate is %f %%'%0.3)
# %를 써서 formatting을 할 때, 우리가 아는 퍼센트 기호는 %%를 써야 나옵니다.
```

```python
print('error rate is 0.3%') # formatting 안 할 때는 그냥 % 기호 쓰면 돼요.
```

```python
print('error rate is %0.3f %%'%0.128309128309124091709)
# 0.xf : 소수 x번째 자리만큼 float type으로 출력해라.
```

```python
print('error rate is %d %%'%0.13948914809)
# formatting에서 d를 넣으라고 시켰기 때문에 그 float의 정수 부분이 나오는 것.
```
